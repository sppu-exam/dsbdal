```python
def fun1(value):
    if (value == "M"):
        return 1
    elif (value == "F"):
        return 2
    else:
        pass
```

```python
df["gender"].value_counts()
```

```python
df["gender"] = df["gender"].apply(fun1)
```

```python
def fun2(value):
    if (np.isnan(value) == False):
        return int(value)
```

```python
df["salary"] = df["salary"].apply(fun2)
```
